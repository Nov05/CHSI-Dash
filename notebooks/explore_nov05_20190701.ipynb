```
# call it in every cell (each is an iframe)
configure_plotly_browser_state()

colorscales = ["Greens", "YlOrRd", "Bluered", "RdBu", "Reds", 
               "Blues", "Picnic", "Rainbow", "Portland", "Jet", 
               "Hot", "Blackbody", "Earth", "Electric", "Viridis", 
               "Cividis"]
portland = [
    [0, 'rgb(12,51,131)'], [0.25, 'rgb(10,136,186)'],
    [0.5, 'rgb(242,211,56)'], [0.75, 'rgb(242,143,56)'],
    [1, 'rgb(217,30,30)']
]
portland_rgb = [i[1] for i in portland]

# log10(population density), age under 19, poverty
x, y, z = (np.log10(demogr[cols[11]].replace([-2222,0], [demogr[cols[11]].mean(),1])), 
           demogr[cols[17]], 
           demogr[cols[14]].replace(-2222.2, demogr[cols[14]].mean()))
d = np.log10(demogr[cols[8]]) # log10(population)

trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=d*2,
        color=z,                     # set color to an array/list of desired values
        colorscale=portland, #colorscales[8],   # choose a colorscale
        opacity=1,
        showscale=True,
        colorbar=dict(x=1, len=0.5, 
                      thickness=10,
                      outlinecolor='white', outlinewidth=0,
                      title=dict(text="Poverty", font=dict(size=10))
                     ),
        line=dict(width=0.001, color='black')
    ),
    text='County Level',
    projection=dict(x=dict(show=True, opacity=0.1, scale=0.4),
                    y=dict(show=True, opacity=0.1, scale=0.4),
                    z=dict(show=True, opacity=0.1, scale=0.4),
                   )
)

data = [trace1]
layout = go.Layout(
    autosize=False,
    width=700,
    height=600,
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    scene=dict(xaxis=dict(title="x = Population Density (lg)"),
               yaxis=dict(title="y = Age Under 19"),
               zaxis=dict(title="z = Poverty"),),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='3d-scatter-colorscale')
```

```
# call it in every cell (each is an iframe)
configure_plotly_browser_state()

# log10(population density), age under 19, poverty
x, y, z = (np.log10(demogr[cols[11]].replace([-2222,0], [demogr[cols[11]].mean(),1])), 
           demogr[cols[17]], 
           demogr[cols[14]].replace(-2222.2, demogr[cols[14]].mean()))
# log10(population)
d = np.log10(demogr[cols[8]]) 

in_range=20

# slicing data
slicenum = 38
slices = np.linspace(0, max(z), slicenum)
if in_range<0 or in_range>slicenum: in_range = 0
slice = slices[in_range]
condition = ((z >= slice) & (z < slices[in_range+1]))
x1 = x[condition]
y1 = y[condition]
z1 = [slice] * len(x1)
slicecolor =cl.interp(portland_rgb, slicenum)[in_range]

# creating a plane
p1 = np.linspace(0, max(x), 5)
p2 = np.linspace(0, max(y), 5)
p1, p2 = np.meshgrid(p1, p2)
p3 = [[slice] * 5] * 5

# this is all data points in very light gray
trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    name='county',
    marker=dict(
        size=d*2,
        color="black", 
        opacity=0.01,
        line=dict(width=0.00, color='black'),
        ),
    showlegend=False,
)

# this is a plane
trace2 = go.Surface(
    x=tuple(p1),
    y=tuple(p2),
    z=tuple(p3),
    name='slice',
    opacity=0.5,
    showlegend=False,
    showscale=False,
)

# this is merely a colorbar
trace3 = go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    name='',
    mode='markers',
    marker=dict(
        size=0.01,
        color=z, # set color to an array/list of desired values
        colorscale=colorscales[8], # choose a colorscale
        opacity=1,
        showscale=True,
        colorbar=dict(x=0.81, len=0.5, 
                      thickness=10,
                      outlinecolor='white', outlinewidth=0,
                      title=dict(text="Poverty", font=dict(size=10))
                     ),
        line=dict(width=0.01, color='gray')
    ),
    showlegend=False,
)

# this is the actual sliced data points
# why separating data points and colorbar?
# cause I need only one color from the colorscale
trace4 = go.Scatter3d(
    x=x1,
    y=y1,
    z=z1,
    name='Poverty',
    mode='markers',
    marker=dict(
        size=d*2,
        color=slicecolor,
        opacity=1,
        showscale=False,
        line=dict(width=0.01, color='gray')
    ),
    projection=dict(x=dict(show=True, opacity=0.1, scale=0.8),
                    y=dict(show=True, opacity=0.1, scale=0.8),
                    z=dict(show=True, opacity=0.1, scale=0.8),
                   ),
    showlegend=False,
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    autosize=True,
    width=1000,
    height=700,
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    scene=dict(xaxis=dict(title="x = Population Density (lg)"),
               yaxis=dict(title="y = Age Under 19"),
               zaxis=dict(title="z = Poverty"),),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='3d-scatter-colorscale')
```

```
# call it in every cell (each is an iframe)
configure_plotly_browser_state()

import plotly as py
import plotly.graph_objs as go
import pandas as pd

url = "https://raw.githubusercontent.com/Nov05/CHSI-Dash/master/data/DEMOGRAPHICS.csv"
demogr = pd.read_csv(url)
cols = demogr.columns.tolist()

# log10(population density), poverty
x = np.log10(demogr[cols[11]].replace([-2222,0], [demogr[cols[11]].mean(),1]))
y = demogr[cols[17]]
z = demogr[cols[14]].replace(-2222.2, demogr[cols[14]].mean())
d = np.log10(demogr[cols[8]]) # log10(population)
  
# create a plane
p1 = np.linspace(0, max(x), 5)
p2 = np.linspace(0, max(y), 5)
p1, p2 = np.meshgrid(p1, p2)
p3 = [[1] * 5] * 5
    
# this is a plane
trace2 = go.Surface(
    x=tuple(p1),
    y=tuple(p2),
    z=tuple(p3),
    name='slice',
    colorscale="Greys",
    opacity=0.5,
### WARNING: This is invalid. It would not cause problem in Colab.
###          But it would fail the plot when deployed on Heroku.
#     showlegend=False,  # invalid property
    showscale=False,
)

data = [trace2]
        
layout = go.Layout(
    title='A Plane',
    autosize=False,
    width=500,
    height=500,
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='3d-surface')
```
