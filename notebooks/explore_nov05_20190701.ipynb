**3D Scatter Plots in Python**  
https://plot.ly/python/3d-scatter-plots/#3d-scatter-plot-with-colorscaling  

```
cols = demogr.columns
pd.DataFrame(demogr.columns).T
```

```
# -2222.2
demogr[cols[14]].value_counts().reset_index().sort_values(by='index', ascending=False)
```

```
pd.DataFrame([np.log10(demogr[cols[8]]), 
              demogr[cols[17]], 
              demogr[cols[14]].replace(-2222.2, 0)
            ]).T.to_csv("3d.csv")
```

```
# then call it in every cell (each is an iframe)
configure_plotly_browser_state()

# log10(population), age under 19, poverty
x, y, z = (np.log10(demogr[cols[8]]), 
           demogr[cols[17]], 
           demogr[cols[14]].replace(-2222.2, 0))

trace1 = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=3,
        color=z,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.3
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig, filename='3d-scatter-colorscale')
```
