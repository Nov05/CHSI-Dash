```
import plotly
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df_LCOD = pd.read_csv('https://raw.githubusercontent.com/Nov05/CHSI-Dash/master/data/LEADINGCAUSESOFDEATH.csv')  
print(df_LCOD.shape)
df_LCOD.sample(5)
```

```
LCOD_columns = np.array(df_LCOD.columns)
CI_cols = [k for k in LCOD_columns if 'CI_' in k]
LCOD_drop = df_LCOD.drop(df_LCOD[CI_cols],axis=1)
LCOD_drop.describe()
```

```
LCOD_clean = LCOD_drop.replace([-1111,-2222],[np.nan,np.nan]) 
LCOD_clean.describe()
```
