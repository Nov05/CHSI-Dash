```
import numpy as np
import pandas as pd
```

```
class CleanData:
  def __init__(self, url):
    self.df = pd.read_csv(url)
    # remove CI from columns
    self.CI_cols = [col for col in np.array(self.df.columns) if 'CI_' in col]
    self.df = self.df.drop(self.df[self.CI_cols],axis=1)
    # replace negative values with NaN
    self.df = self.df.replace([-1,-1111,-1111.1,-2,-2222.2,-2222,-9999,-9989.9],
                              [np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,
                               np.nan])
  def dataframe(self):
     return self.df
```

```
url = 'https://raw.githubusercontent.com/Nov05/CHSI-Dash/master/data/LEADINGCAUSESOFDEATH.csv'
df_LCOD = CleanData(url).dataframe()
print(df_LCOD.shape)
df_LCOD.sample(5)
```
