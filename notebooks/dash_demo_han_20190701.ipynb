```python
cod_cols = cod.columns.values
cod_cols = [c for c in cod_cols if 'CI' not in c]
print(cod_cols)
```

```python
cod['State_FIPS_Code'] = cod['State_FIPS_Code'].apply(lambda x: str(x).zfill(2))
cod['County_FIPS_Code'] = cod['County_FIPS_Code'].apply(lambda x:str(x).zfill(3))
cod['FIPS'] = cod['State_FIPS_Code'] + cod['County_FIPS_Code']

cod['A_Wh_Comp'] = cod['A_Wh_Comp'].replace([-1111, -2222], [np.nan, np.nan])

colorscale = ["#f7fbff", "#ebf3fb", "#deebf7", "#d2e3f3", "#c6dbef", "#b3d2e9", "#9ecae1",
    "#85bcdb", "#6baed6", "#57a0ce", "#4292c6", "#3082be", "#2171b5", "#1361a9",
    "#08519c", "#0b4083", "#08306b"
]
endpts = list(np.linspace(1, 100, len(colorscale) - 1))

fips = cod['FIPS'].tolist()
values = cod['A_Wh_Comp'].tolist()
```

```python
%%time
fig = ff.create_choropleth(
    fips = fips, values = values, scope = ['usa'],
    binning_endpoints = endpts, colorscale = colorscale,
    show_state_data = False,
    show_hover = True, centroid_marker = {
        'opacity': 0
    },
    asp = 2.9,
    title = 'USA by A_Wh_Comp',
    legend_title = 'A_Wh_Comp'
)
py.iplot(fig, filename = 'choropleth_full_usa')
```

```python
tx_cod = cod.loc[cod['CHSI_State_Name'] == "Texas"]
tx_fips = tx_cod['FIPS'].tolist()
tx_values = tx_cod['A_Wh_Comp'].tolist()
```

```python
%%time
fig2 = ff.create_choropleth(
    fips = tx_fips, values = tx_values, scope = ['Texas'],
    binning_endpoints = endpts, 
    colorscale = colorscale,
    show_state_data = False,
    show_hover = True, centroid_marker = {
        'opacity': 0
    },
    asp = 2.9,
    title = 'Everything Is Bigger In Texas',
    legend_title = 'A_Wh_Comp'
)
py.iplot(fig2, filename = 'texas')
```
